#### Calculando FFT e PSD

````python
dt = 0.001
tamanho = len(tempo)                                             # Tamanho do vetor tempo
resultado_fft = np.fft.fft(sinal, tamanho)                       # Calcula o FFT
resultado_psd = resultado_fft * np.conj(resultado_fft) / tamanho # Power spectrum (power por freq)
x_freq = ( 1 / ( dt * tamanho )) * np.arange(tamanho)            # Cria uma lista com frequencias para usarmos no eixo x

indices = resultado_psd > 100                                   # Mascara booleana
resultado_ffilt = np.fft.ifft(resultado_fft * indices)          # Filtra o sinal
````

````python
fig, ax = plt.subplots(nrows=3, ncols=1, figsize=(25,8))

# Compara sinal com ruído e filtrado pelo FFT
ax[0].plot( x_freq, sinal, color='r', label='Original' )
ax[0].plot( x_freq, resultado_ffilt, color='b', label='Filtrado' )
ax[0].set_title('Sinal Original x Filtrado')
ax[0].legend()

# Compara sinal original, sem ruído com o filtrado pelo fft
ax[1].plot( x_freq, data.combinado_sr.to_list(), color='r', label='Original sem ruído' )
ax[1].plot( x_freq, resultado_ffilt, color='b', label='Filtrado' )
ax[1].set_title('Sinal Original sem ruído x Filtrado')
ax[1].legend()

# Mostra o PSD encontrado
ax[2].plot( x_freq, resultado_psd, color='r', label="Com ruído" )
ax[2].plot( x_freq, resultado_psd * indices, color='b', label="Filtrado" )
ax[2].set_xlim( 0, 80 )
ax[2].set_title('PSD normal x PSD sem ruído')
ax[2].legend()

plt.show()
````

````python
def gera_grafico( numero, eixo_x, eixo_y, cor, label, titulo = '', mostra_titulo = False, mostra_legenda = False, spines = [True, True, True, True]):
  """
    Essa função é usada para criar gráficos. Recebe como entrada um numero de figura, 
    valores para eixos X e Y, cor, descrição de legenda, um título e dois booleanos
    que determinam se o título e a legenda serão mostradas ou não, assim como uma lista
    para os spines
  """
  ax[numero].plot( eixo_x, eixo_y, color=cor, label=label )
  # Remove spines
  ax[numero].spines['left'].set_visible(spines[0])
  ax[numero].spines['bottom'].set_visible(spines[1])
  ax[numero].spines['right'].set_visible(spines[2])
  ax[numero].spines['top'].set_visible(spines[3])

  # Mostra ou não o título
  if mostra_titulo == True:
    ax[numero].set_title(titulo)
  
  # Mostra ou não a legenda
  if mostra_legenda == True:
    ax[numero].legend()

fig, ax = plt.subplots(nrows=3, ncols=1, figsize=(25,8))

# Compara sinal com ruído e filtrado pelo FFT
gera_grafico( 0, x_freq, sinal, 'red', 'Original', '', False, False, [False, False, False, False])
gera_grafico( 0, x_freq, resultado_ffilt, 'blue', 'Filtrado', 'Sinal Original x Filtrado', True, True, [False, False, False, False])

# Compara sinal original, sem ruído com o filtrado pelo fft
gera_grafico( 1, x_freq, data.combinado_sr.to_list(), 'r', 'Original sem ruído')
gera_grafico( 1, x_freq, resultado_ffilt, 'b', 'Filtrado', 'Sinal Original sem ruído x Filtrado', True, False, [False, False, False, False])

# Mostra o PSD encontrado
gera_grafico( 2, x_freq, resultado_psd, 'r', 'Com ruído')
gera_grafico( 2, x_freq, resultado_psd * indices, 'b', 'Filtrado', 'PSD normal x PSD sem ruído', 0, 1, [0, 1, 0, 0])
ax[2].set_xlim( 0, 80 )
````

## 3.4 Análise

````python
sinal_filtrado = filter_signal(sinal, fs, 'lowpass', limiar, n_seconds=tam_filtro_s, remove_edges=False, plot_properties=True)
````

Encontrando picos e vales.

````python
from bycycle.cyclepoints import find_extrema, find_zerox
from bycycle.plts import plot_cyclepoints_array
from bycycle.features import compute_features
````

````python
n_seconds = .75
band = (2, 30)
peaks, troughs = find_extrema(sinal_filtrado, fs, band, filter_kwargs={'n_seconds':n_seconds})
````

````python
rises, decays = find_zerox(sinal_filtrado, peaks, troughs)

plot_cyclepoints_array(sinal_filtrado, fs, peaks=peaks, troughs=troughs, rises=rises, decays=decays)
````

Computando características
